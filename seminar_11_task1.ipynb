### Дана функция f = 2 * x ** 3 + 2 * x ** 2 - 18 * x - 18

```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
f = 2 * x ** 3 + 2 * x ** 2 - 18 * x - 18
f
```

### 1. Определить корни

```python
solveset (f,x)
```

### 2. Найти интервалы, на которых функция возрастает
### 3. Найти интервалы, на которых функция убывает

```python
b=[-oo, oo]
b[1:1] = solve(diff(f), x)
b
```

```python
c=[]
d=[]
for i in range(1, len(b)):
    boo=is_increasing(f, Interval.open(b[i-1], b[i]))
    if boo:
        c.append(f"[{round(b[i-1], 2)}, {round(b[i], 2)}]")
    else:
        d.append(f"[{round(b[i-1], 2)}, {round(b[i], 2)}]")
        
print(f"Возрастает на интервалах:", *c, sep="\n")
print(f"Убывает на интервалах:", *d, sep="\n")
```

### 4. Построить график

```python
plot(f, (x, -3.5, 3.5), legend=True)
```

### 5. Вычислить вершину
### Экстремумы функции

```python
e=solve(diff(f),x)
for i in e:
    g = f.subs(x,i)
    if g<0:
        print(f'Точка минимума: x:{round(i, 2)} y:{round(g, 2)}')
    elif g>0:
        print(f'Точка максимума: x:{round(i, 2)} y:{round(g, 2)}')
```

### 6. Определить промежутки, на котором f > 0
### 7. Определить промежутки, на котором f < 0

```python
m=[-oo, oo]
incr_list=[]
decr_list=[]
m[1:1]=solve(f,x)
for i in range(1, len(m)):
    boo=is_increasing(f, Interval.open(m[i-1], m[i]))
    if boo:
        incr_list.append(f"[{m[i-1]}, {m[i]}]")
    else:
        decr_list.append(f"[{m[i-1]}, {m[i]}]")
        
print("f > 0:", *incr_list, sep="\n")
print("f < 0:", *decr_list, sep="\n")
```
